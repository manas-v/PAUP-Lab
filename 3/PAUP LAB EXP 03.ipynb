## Scikit-learn Design

Primarily three types of objects:


1.Estimator-->it estimates some parameter based on dataset.for ex-imputer
              it has a fit method and transform method
              fit method-fits the dataset and calculates internal parameters
              
              
2.Transformer-->transform method takes input and returns output based on the 
                 leartnings of fit() method,also has a convenience method called
                 fit_transform() which fits and then transforms
                 
                 
3.Predictor-->LinearRegressionmodel() is an example of predictor.this has fit and predict 
              and also has a score function to evaluate the prediction

# Feature Scaling

Two types of scaling methods are there:
    
    1.Min-max scaler(Normalization)
    (value-min)/(max-min)
    sklearn provides a class called MinMaxScaler for this
    
    2.standardized scaler
      (value-mean)/sd
      here sklearn provides a class called StandardScaler
